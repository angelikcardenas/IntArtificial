# N-Queens

```python
def queensproblem(filas, columnas):
    soluciones = [[]]
    for fila in range(filas):
        soluciones = add_one_queen(fila, columnas, soluciones)
    return soluciones

def add_one_queen(n_fila, columnas, prev_soluciones):
    return [solucion + [n_columna]
            for solucion in prev_soluciones
            for n_columna in range(columnas)
            if no_conflict(n_fila, n_columna, solucion)]

def no_conflict(n_fila, n_columna, solucion):
    return all(solucion[fila]       != n_columna           and
               solucion[fila] + fila != n_columna + n_fila and
               solucion[fila] - fila != n_columna - n_fila
               for fila in range(n_fila))

count = 0
```

```python
r=c= input("Ingrese numero de filas y columnas: ")
for solucion in queensproblem(int(r), int(c)):
    count += 1
    print(solucion)

print "Total de soluciones posibles: " ,count
```
